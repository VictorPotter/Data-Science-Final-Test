```python
# Criação da estrutura da rede neural com a classe Sequential (sequência de camadas)
from keras.models import Sequential
from keras.layers import Dense, Dropout

modelo_deep = Sequential()
#primeira camada oculta
modelo_deep.add(Dense(units = 20, activation = 'relu', input_dim = 19))
modelo_deep.add(Dropout(0.2))
#segunda camada oculta
modelo_deep.add(Dense(units = 8, activation = 'relu'))
modelo_deep.add(Dropout(0.2))
#terceira camada oculta
modelo_deep.add(Dense(units = 8, activation = 'relu'))
modelo_deep.add(Dropout(0.2))

# Função softmax para quando temos um problema de classificação com mais de duas classes 
#(é gerada uma probabilidade em cada neurônio)
modelo_deep.add(Dense(units = 2, activation = 'sigmoid'))
#São 2 neuronios de saída pq são 2 possibilidades de classificação de classe (Aprovado ou reprovado)
```

```python
# Visualização da estrutura da rede neural
modelo_deep.summary()
```

```python
# Configuração dos parâmetros da rede neural (adam = algoritmo para atualizar os pesos e loss = cálculo do erro)
modelo_deep.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])

# Treinamento, dividindo a base de treinamento em uma porção para validação (validation_data)
historico = modelo_deep.fit(X_trein_deep, y_trein_deep, batch_size = 10, epochs = 100, validation_data = (X_teste_deep, y_teste_deep))
```

```python
# Gráfico para visualizar os erros e accuracy
historico.history.keys()
#evolução do erro, azul
plt.plot(historico.history['val_loss'])
#performance da rede
plt.plot(historico.history['val_accuracy'])
```
